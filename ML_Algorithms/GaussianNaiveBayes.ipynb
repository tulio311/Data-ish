```python
import numpy as np

# Inciso (c)

# Este es el clasificador encontrado
def g(k):
  if k < 13:
    return 1
  elif k < 18:
    return 2
  else:
    return 3

# Número de pruebas que se harán
m = 1000000

# Aquí guardaremos la cantidad de datos bien clasificados
sum = 0

for i in range(m):
  # Se elige con igual probabilidad una de las 3 familias y luego se guarda el
  # lambda respectivo
  a = np.random.randint(1,4)
  if a == 1:
    Lambda = 10
  elif a==2:
    Lambda = 15
  else:
    Lambda = 20

  x = np.random.poisson(Lambda)

  # Si el dato se clasifica bien, se suma 1 a sum
  if g(x) == a:
    sum +=1

# Tasa de error
print(1 - sum/m)

# El resultado si se corresponde con el error teórico
```

```python
# Ejercicio 2

import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
import pandas as pd

data = pd.read_csv("wine.data",header=None)

# Se extrae el vector de etiquetas y se acomoda para el modelo
y = data.iloc[0:,[0]]
y = y.transpose().to_numpy()[0]

# Se extraen las características y se acomodan para el modelo
X = data.iloc[0:,1:].to_numpy()

# Se define y entrena el modelo
model = GaussianNB()
model.fit(X, y)

# Se verifican sus predicciones en los datos
Pred = model.predict(X)

# Se cuentan las fallas y se imprimen
sum = 0
for i in range(len(y)):
  if y[i] != Pred[i]:
    sum += 1

print(sum)

```
