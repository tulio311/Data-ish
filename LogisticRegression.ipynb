```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
iris = datasets.load_iris()
X = iris.data[: , :2]
Y = iris.target
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.linear_model import LogisticRegression
```

```python
# Se pone la variable multi_class='ovr' indicando que se usa el modo one-vs-rest, que es el algoritmo
# descrito en la tarea.

model = LogisticRegression(multi_class='ovr')
clasif = model.fit(X,Y)
```

```python
Y_pred = model.predict(X)

# Checamos los que se clasificaron mal
a = [Y == Y_pred]
sum = 0
for x in a[0]:
  if x == False:
    sum +=1

# Imprimimos el número de datos mal clasificados
print(sum)
```

```python
# Graficamos
disp = DecisionBoundaryDisplay.from_estimator(clasif, X, response_method="predict")
disp.ax_.scatter(X[:, 0], X[:, 1], c=iris.target, edgecolor="k")
plt.show()
```
